```python
import pandas as pd
```

```python
df = pd.read_csv('/content/results.csv')
display(df.head())
```

```python
# Convert 'date' column to datetime objects
df['date'] = pd.to_datetime(df['date'])
```

```python
# Filter the DataFrame to keep rows where the year is between 2023 and 2026
df_filtered = df[(df['date'].dt.year >= 2023) & (df['date'].dt.year <= 2026)]

display(df_filtered.head())
```

```python
df_filtered = df_filtered.drop(columns=['city', 'country', 'neutral'])
display(df_filtered.head())
```

```python
def get_winner(row):
    if row['home_score'] > row['away_score']:
        return row['home_team']
    elif row['away_score'] > row['home_score']:
        return row['away_team']
    else:
        return 'Draw'

df_filtered['winner'] = df_filtered.apply(get_winner, axis=1)
display(df_filtered.head())
```
