```python
import pandas as pd

file_path = '/content/whoscored_player_stats_passing.csv'
df = pd.read_csv(file_path)
display(df.head())
```

```python
pd.set_option('display.max_columns', None)
display(df.head())
```

```python
columns_to_drop = [
    'height',
    'weight',
    'age',
    'isManOfTheMatch',
    'isActive',
    'playedPositions',
    'playedPositionsShort',
    'teamRegionName',
    'regionCode',
    'tournamentShortName',
    'subOn',
    'name',
    'firstName',
    'lastName',
    'playerId',
    'teamId',
    'teamName',
    'seasonId',
    'seasonName',
    'isOpta',
    'tournamentId',
    'tournamentRegionId',
    'tournamentRegionCode',
    'tournamentRegionName',
    'tournamentName',
    'minsPlayed',
    'ranking'
]

df = df.drop(columns=columns_to_drop)
display(df.head())
```

```python
aggregation_functions = {
    'apps': 'sum',
    'assistTotal': 'sum',
    'keyPassPerGame': lambda x: x[x != 0].mean(),
    'totalPassesPerGame': lambda x: x[x != 0].mean(),
    'accurateCrossesPerGame': lambda x: x[x != 0].mean(),
    'accurateLongPassPerGame':lambda x: x[x != 0].mean(),
    'accurateThroughBallPerGame': lambda x: x[x != 0].mean(),
    'rating': lambda x: x[x != 0].mean(),
    'passSuccess': lambda x: x[x != 0].mean()
}

df_agg = df.groupby(['Player_ID', 'Team_Name', 'Player_Name']).agg(aggregation_functions).reset_index()
display(df_agg.head())
```

```python
from google.colab import files

output_filename = 'whoscored_player_stats_passing_processed.csv'
df_agg.to_csv(output_filename, index=False)
files.download(output_filename)
```
