```python
import pandas as pd


pd.set_option('display.max_columns', None)
df = pd.read_csv('/content/whoscored_player_stats_defensive.csv')
display(df.head())
```

```python
columns_to_drop = [
    'height', 'weight', 'age', 'isManOfTheMatch', 'isActive',
    'playedPositions', 'playedPositionsShort', 'teamRegionName', 'regionCode',
    'tournamentShortName', 'subOn', 'name', 'firstName', 'lastName',
    'playerId', 'teamId', 'teamName', 'seasonId', 'seasonName', 'isOpta',
    'tournamentId', 'tournamentRegionId', 'tournamentRegionCode',
    'tournamentRegionName', 'tournamentName', 'rating', 'minsPlayed', 'ranking'
]

df_dropped = df.drop(columns=columns_to_drop)

# Display the first 5 rows of the DataFrame with the dropped columns
display(df_dropped.head())
```

```python
aggregation_functions = {
    'apps': 'sum',
    'goalOwn': 'sum',
    'tacklePerGame': lambda x: x[x != 0].mean(),
    'interceptionPerGame': lambda x: x[x != 0].mean(),
    'foulsPerGame': lambda x: x[x != 0].mean(),
    'offsideWonPerGame': lambda x: x[x != 0].mean(),
    'clearancePerGame': lambda x: x[x != 0].mean(),
    'wasDribbledPerGame': lambda x: x[x != 0].mean(),
    'outfielderBlockPerGame': lambda x: x[x != 0].mean()
}

df_aggregated = df.groupby(['Player_ID', 'Player_Name']).agg(aggregation_functions).reset_index()
display(df_aggregated.head())
```

```python
df_aggregated.to_csv('aggregated_player_stats.csv', index=False)
```
