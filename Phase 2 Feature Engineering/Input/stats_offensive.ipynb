```python
import pandas as pd

file_path = '/content/whoscored_player_stats_offensive.csv'
df = pd.read_csv(file_path)
display(df.head())
```

```python
pd.set_option('display.max_columns', None)
display(df.head())
```

```python
columns_to_drop = [
    'height',
    'weight',
    'age',
    'isManOfTheMatch',
    'isActive',
    'playedPositions',
    'playedPositionsShort',
    'teamRegionName',
    'regionCode',
    'tournamentShortName',
    'subOn',
    'name',
    'firstName',
    'lastName',
    'playerId',
    'positionText',
    'teamId',
    'teamName',
    'seasonId',
    'seasonName',
    'isOpta',
    'tournamentId',
    'tournamentRegionId',
    'tournamentRegionCode',
    'tournamentRegionName',
    'tournamentName',
    'minsPlayed',
    'ranking'
]

df = df.drop(columns=columns_to_drop, errors='ignore')
display(df.head())
```

```python
aggregation_functions = {
    'apps': 'sum',
    'goal': 'sum',
    'assistTotal': 'sum',
    'shotsPerGame': lambda x: x[x != 0].mean(),
    'keyPassPerGame': lambda x: x[x != 0].mean(),
    'dribbleWonPerGame': lambda x: x[x != 0].mean(),
    'foulGivenPerGame': lambda x: x[x != 0].mean(),
    'offsideGivenPerGame': lambda x: x[x != 0].mean(),
    'dispossessedPerGame': lambda x: x[x != 0].mean(),
    'turnoverPerGame': lambda x: x[x != 0].mean(),
    'rating': lambda x: x[x != 0].mean()
}

df_aggregated = df.groupby(['Player_ID', 'Team_Name', 'Player_Name']).agg(aggregation_functions).reset_index()
display(df_aggregated.head())
```

```python
output_file_name = 'whoscored_player_stats_offensive_aggregated.csv'
df_aggregated.to_csv(output_file_name, index=False)

from google.colab import files
files.download(output_file_name)
```
