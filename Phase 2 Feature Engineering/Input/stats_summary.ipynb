```python
import pandas as pd
```

```python
df = pd.read_csv('/content/whoscored_player_stats.csv')
display(df.head())
```

```python
pd.set_option('display.max_columns', None)
display(df.head())
```

```python
columns_to_drop = [
    'height', 'weight', 'age', 'isManOfTheMatch', 'isActive', 'minsPlayed',
    'playedPositions', 'playedPositionsShort', 'teamRegionName',
    'regionCode', 'tournamentShortName', 'subOn', 'name',
    'firstName', 'lastName', 'playerId', 'teamId', 'teamName',
    'seasonId', 'seasonName', 'isOpta', 'tournamentId',
    'tournamentRegionId', 'tournamentRegionCode', 'tournamentRegionName',
    'tournamentName', 'ranking'
]
df = df.drop(columns=columns_to_drop)
display(df.head())
```

```python
aggregation_functions = {
    'apps': 'sum',
    'manOfTheMatch': 'sum',
    'goal': 'sum',
    'assistTotal': 'sum',
    'yellowCard': 'sum',
    'redCard': 'sum',
    'shotsPerGame': lambda x: x[x != 0].mean(),
    'aerialWonPerGame': lambda x: x[x != 0].mean(),
    'passSuccess': lambda x: x[x != 0].mean(),
    'rating': lambda x: x[x != 0].mean()
}

df_aggregated = df.groupby(['Player_ID', 'Player_Name']).agg(aggregation_functions).reset_index()
display(df_aggregated.head())
```

```python
df_aggregated.to_csv('df_aggregated.csv', index=False)
print('df_aggregated.csv has been downloaded.')
```
